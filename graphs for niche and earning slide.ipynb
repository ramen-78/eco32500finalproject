```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('sales each month.csv')

months = df['Order_Month']
avg_benefit_per_order = df['Avg_Benefit_per_order']

plt.figure(figsize=(10, 6))
plt.plot(months, avg_benefit_per_order, marker='o', linestyle='-', color='orange', label='Avg Benefit per Order')
plt.xlabel('Month')
plt.ylabel('Avg Benefit per Order')
plt.title('Average Benefit per Order by Month')
plt.xticks(months)
plt.legend()

coefficients = np.polyfit(months, avg_benefit_per_order, 1)
trend_line = np.polyval(coefficients, months)

plt.plot(months, trend_line, linestyle='--', color='blue', label='Trend Line')

plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('category with most late delivery.csv')

# Extract category names and corresponding number of late deliveries
category_names = df['Category_Name']
late_deliveries = df['Late_Deliveries']

# Create a pie chart
plt.figure(figsize=(10, 8))
plt.pie(late_deliveries, labels=category_names, autopct='%1.1f%%', startangle=140)
plt.title('Late Deliveries by Category')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
