## Azure SQL Query for Year, Month and Total_Sales

SELECT

    YEAR(OD.order_date_DateOrders) AS Year,
    MONTH(OD.order_date_DateOrders) AS Month,
    SUM(OD.Sales) AS Total_Sales
FROM

    OrderData OD
GROUP BY

    YEAR(OD.order_date_DateOrders),
    MONTH(OD.order_date_DateOrders)
ORDER BY

    Month, Year;


## Google Colab Module for Total Sales By Year and Month Represented in a Line-PLot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset for total sales
df_sales = pd.read_csv("Y_M_S.csv")

# Define month names
month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Map month values to month names
df_sales['Month'] = df_sales['Month'].map(month_names)

# Sort the DataFrame by month names and years
df_sales['Month'] = pd.Categorical(df_sales['Month'], categories=month_names.values(), ordered=True)
df_sales = df_sales.sort_values(['Month', 'Year'])

# Define custom palette
custom_palette = sns.color_palette("husl", len(df_sales['Year'].unique()))

# Format numbers with commas
def format_number(num):
    return f"{num:,.2f}"

# Create line-plot for total sales
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_sales, x='Month', y='Total_Sales', hue='Year', marker='o', palette=custom_palette)
plt.title('Total Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

# Calculate total sales for each year
total_sales_per_year = df_sales.groupby('Year')['Total_Sales'].sum()

# Add total sales to the legend
handles, labels = plt.gca().get_legend_handles_labels()
for year, total_sales in total_sales_per_year.items():
    labels[labels.index(str(year))] += f' (Total Sales: {format_number(total_sales)})'

# Create legend for total sales
plt.legend(handles, labels, title='Total Sales', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

## Google Colab Module for Total Sales By Year and Month Represented in a Heat-Map

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset for total sales
df_sales = pd.read_csv("Y_M_S.csv")

# Define month names
month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Map the month values to month names
df_sales['Month'] = df_sales['Month'].map(month_names)

# Sort the DataFrame by month names and years
df_sales['Month'] = pd.Categorical(df_sales['Month'], categories=month_names.values(), ordered=True)
df_sales = df_sales.sort_values(['Month', 'Year'])

# Pivot the data to create a matrix of total sales for each year and month
pivot_sales = df_sales.pivot_table(index='Month', columns='Year', values='Total_Sales', aggfunc='sum')

# Plotting the heat-map for total sales
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_sales, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=0.5, linecolor='black')

plt.title('Total Sales by Year and Month')
plt.xlabel('Year')
plt.ylabel('Month')

plt.tight_layout()
plt.show()
```

## Azure SQL Query for Year, Month and Profits

SELECT

    YEAR(OD.order_date_DateOrders) AS Year,
    MONTH(OD.order_date_DateOrders) AS Month,
    SUM(OD.Order_Item_Profit_Ratio) AS Profits
FROM

    OrderData OD
GROUP BY

    YEAR(OD.order_date_DateOrders),
    MONTH(OD.order_date_DateOrders)
ORDER BY

    Month, Year;
